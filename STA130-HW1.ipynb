7. df.dropna() removes a row with missing data, while it keeps other rows intact, while del df['col'] removes an entire column that contains missing data. df.dropna() is more efficient when specific rows want to be removed. If a column contains few missing data, and the rest of the column needs to be kept, it is better to use df.dropna() since it keeps the rest of the dataset intact. If there is a column that contains a lot of missing data, df.dropna() will delete too much data which is probably not what we want. del df['col'] is more efficient when there is a column that contains lots of missing data, or there is an entire column that is no longer relevant regardless of its content. In these cases, it will be faster to just delete the entire column. It is also better to apply df.dropna() after removing an entire column with too many missing data, since df.dropna() will then not remove too many data that may still be useful.
